#1.What is hypothesis testing in statistics?
 - Hypothesis testing is to test some hypothesis about parent population from
   which the sample is drawn.

#2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis is the claim that there's no effect in the population.
     the null hypothesis of a test always predicts no effect or no relationship
     between variables but in alternative hypothesis there is an effect, a difference occured.

#3.What is the significance level in hypothesis testing, and why is it important?
  - Level of significance is the measurement of the statistical significance.
     The level of significance explains whether the null hypothesis is accepted or rejected and denoted by alpha it is important for Controls the rate of false positives (Type I error), guides decision-making, Consistency in testing.

#4.What does a P-value represent in hypothesis testing?
  - The p-value, or probability value, is a statistical measure used in
     hypothesis testing to assess the strength of evidence against a null hypothesis.

#5.How do you interpret the P-value in hypothesis testing?
  - To interpret the p-value, you need to compare it to a chosen significance
     level (alpha). During hypothesis testing, we assume a significance level,

     if Small P-value (≤ α)-Strong evidence against the null hypothesis → Reject H₀

     Large P-value (> α)-Weak evidence against the null → Fail to reject H₀

     P-value ~ α - Results are borderline significant

#6.What are Type 1 and Type 2 errors in hypothesis testing?
  - Type 1 error is rejection of the null hypothesis it is actually true.
     e.g- An innocent persion is convicted.

  - Type 2 error is the failure to reject the null hypothesis that is actually  false.
     e.g- A guilty persion may not be convicted.

#7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - One and Two-Tailed Tests are ways to identify the relationship between the
   statistical variables. For checking the relationship between variables in a single direction (Left or Right direction), we use a one-tailed test. A two-tailed test is used to check whether the relations between variables are in any direction or not.

#8.What is the Z-test, and when is it used in hypothesis testing?
  - The Z-test is a statistical test used to determine whether there is a
     significant difference between sample and population means when the population standard deviation is known and the sample size is large (typically ≥ 30).

  - It is used when large sample size (n ≥ 30) and population of standard
     deviation (σ) is known.

#9.How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - The Z-score is calculated using the formula: Z = (X - μ) / σ, where X is  
     the raw data point, μ is the population mean, and σ is the standard deviation. When the mean and standard deviation of the population are unknown, use the sample mean (¯X) and standard deviation (S) instead.

  - Calculate the Z-score of your sample statistic.

  - Compare it to a critical Z-value from the Z-distribution (based on your     significance level, e.g., 1.96 for 5% significance in a two-tailed test).

  - If the Z-score falls in the rejection region, you reject the null
     hypothesis.

#10.What is the T-distribution, and when should it be used instead of the normal distribution?
   - The t-distributon,also known as the Student's t-distribution, is a
     probability distribution that is used to estimate the mean of a normally distributed population when the sample size is small or the population standard deviation is unknown. the t-distribution is used instead of the normal distribution can greatly enhance the accuracy of our statistical analyses.

#11.What is the difference between a Z-test and a T-test?
   -  Z-tests are used when the population variance is known and the sample
      size is large, while t-tests are used when the population variance is unknown and the sample size is small.

#12.What is the T-test, and how is it used in hypothesis testing?
   - A t test is a statistical test that is used to compare the means of two
     groups. It is often used in hypothesis testing to determine whether a process or treatment actually has an effect on the population of interest, or whether two groups are different from one another.

#13.What is the relationship between Z-test and T-test in hypothesis testing?
   - The Z-test and T-test are both statistical methods used to test
     hypotheses about population means, but they are used under different conditions.
   - Both are used to determine if there is a significant difference between
     sample and population means (or between two sample means).

#14.What is a confidence interval, and how is it used to interpret statistical results?
   - A confidence interval (CI) is a range of values that is likely to  
     contain the true population parameter(like the mean or proportion) with a certain level of confidence (e.g., 95%).

     Interpretation:

   - 95% confidence interval for the average height of students in a class  
     and get: (165cm, 175cm)

   - We are 95% confident that the true average height lies between 165 cm  
     and 175 cm.

  - It does not mean there's a 95% chance that the true value is in this  
     range — the true value is fixed; the interval varies from sample to sample.

#15.What is the margin of error, and how does it affect the confidence interval?
   - A margin of error tells you how many percentage points your results will
     differ from the real population value. For example, a 95% confidence interval with a 4 percent margin of error means that your statistic will be within 4 percentage points of the real population value 95% of the time.It affect confidence level,sample variability,sample size.

```python
#3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std):
    sample = np.array(sample)
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    return z_score, p_value

# Example usage:
sample_data = [101, 98, 100, 97, 102, 99, 100, 98]
pop_mean = 100
pop_std = 2

z, p = one_sample_z_test(sample_data, pop_mean, pop_std)
print(f"Z-score: {z:.4f}")
print(f"P-value: {p:.4f}")

# Interpretation
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
mu_0 = 100
sigma = 15
n = 30
alpha = 0.05

np.random.seed(0)
sample = np.random.normal(loc=105, scale=15, size=n)
sample_mean = np.mean(sample)

# Z-test statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Critical values
z_crit = stats.norm.ppf(1 - alpha / 2)
lower_crit = mu_0 - z_crit * (sigma / np.sqrt(n))
upper_crit = mu_0 + z_crit * (sigma / np.sqrt(n))

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical values: [{lower_crit:.2f}, {upper_crit:.2f}]")
print("Result:", "Reject H0" if abs(z) > z_crit else "Fail to reject H0")

# Visualization
x = np.linspace(mu_0 - 4, mu_0 + 4, 1000)
pdf = stats.norm.pdf(x, mu_0, sigma / np.sqrt(n))

plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='Null Distribution (H₀)', color='blue')
plt.axvline(lower_crit, color='red', linestyle='--', label='Critical Region')
plt.axvline(upper_crit, color='red', linestyle='--')
plt.axvline(sample_mean, color='green', linestyle='-', label='Sample Mean')

plt.fill_between(x, 0, pdf, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.3)
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualize Type I and Type II errors for hypothesis test.
    mu0: mean under H0
    mu1: mean under H1
    sigma: population std deviation
    n: sample size
    alpha: significance level
    """
    se = sigma / np.sqrt(n)
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions
    h0_pdf = norm.pdf(x, mu0, se)
    h1_pdf = norm.pdf(x, mu1, se)

    # Critical value for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)
    lower = mu0 - z_crit * se
    upper = mu0 + z_crit * se

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(x, h0_pdf, label='H₀ distribution', color='blue')
    plt.plot(x, h1_pdf, label='H₁ distribution', color='red')

    # Type I Error (rejection region under H0)
    plt.fill_between(x, 0, h0_pdf, where=(x < lower) | (x > upper), color='blue', alpha=0.3, label='Type I Error (α)')

    # Type II Error (accept region under H1)
    plt.fill_between(x, 0, h1_pdf, where=(x >= lower) & (x <= upper), color='red', alpha=0.3, label='Type II Error (β)')

    # Labels
    plt.axvline(lower, color='black', linestyle='--')
    plt.axvline(upper, color='black', linestyle='--')
    plt.title("Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage
plot_type1_type2(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results.
from scipy import stats
import numpy as np

def perform_independent_ttest(group1, group2, alpha=0.05, equal_var=True):

    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=equal_var)

    interpretation = "Fail to reject the null hypothesis: No significant difference between the groups."
    if p_val < alpha:
        interpretation = "Reject the null hypothesis: Significant difference between the groups."

    return {
        't_statistic': t_stat,
        'p_value': p_val,
        'alpha': alpha,
        'interpretation': interpretation
    }

# Example run
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=105, scale=10, size=30)

perform_independent_ttest(group1, group2, alpha=0.05, equal_var=True)

```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Sample paired data
before = [85, 90, 88, 75, 95, 92, 89, 94, 91, 87]
after = [88, 91, 90, 78, 96, 94, 91, 96, 93, 89]

# Paired T-test
t_stat, p_val = stats.ttest_rel(before, after)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("Significant difference (reject H0)")
else:
    print("No significant difference (fail to reject H0)")

# Visualization
plt.figure(figsize=(8, 5))
plt.plot(before, label='Before', marker='o')
plt.plot(after, label='After', marker='o')
plt.title('Before vs After Paired Sample')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
#8.Simulate data and perform both Z-test and T-test, then compare the results using Python.
import numpy as np
from scipy import stats

# Step 1: Simulate data
np.random.seed(42)
sample1 = np.random.normal(loc=100, scale=15, size=30)  # population mean = 100
sample2 = np.random.normal(loc=105, scale=15, size=30)  # mean shifted

# Step 2: Z-test (assume known population std dev = 15)
pop_std = 15
z_stat = (np.mean(sample1) - np.mean(sample2)) / np.sqrt((pop_std**2 / len(sample1)) + (pop_std**2 / len(sample2)))
p_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Step 3: T-test (scipy uses Welch's t-test by default)
t_stat, p_t = stats.ttest_ind(sample1, sample2, equal_var=False)

# Step 4: Print and compare results
print("=== Z-test ===")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_z:.4f}")

print("\n=== T-test ===")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_t:.4f}")

# Simple interpretation
alpha = 0.05
print("\nZ-test result:", "Reject H0" if p_z < alpha else "Fail to reject H0")
print("T-test result:", "Reject H0" if p_t < alpha else "Fail to reject H0")
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):

    n = len(data)
    mean = np.mean(data)
    se = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * se
    return (mean - margin, mean + margin)

# Example usage
np.random.seed(0)
sample = np.random.normal(loc=100, scale=15, size=30)
ci = confidence_interval(sample, confidence=0.95)
print(f"95% Confidence Interval: {ci}")
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(data, confidence=0.95):

    n = len(data)
    se = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)
    moe = t_crit * se
    return moe

# Example usage
np.random.seed(1)
sample = np.random.normal(loc=50, scale=10, size=25)
moe = calculate_margin_of_error(sample, confidence=0.95)
print(f"Margin of Error (95% CI): {moe:.2f}")
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, likelihood, marginal):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(H), prior probability of hypothesis
    - likelihood: P(D|H), probability of data given hypothesis
    - marginal: P(D), probability of data (evidence)

    Returns:
    - posterior: P(H|D)
    """
    posterior = (likelihood * prior) / marginal
    return posterior

# Example values
P_H = 0.01
P_D_given_H = 0.99
P_D_given_not_H = 0.05
P_not_H = 1 - P_H

# Total probability of positive test (evidence)
P_D = (P_D_given_H * P_H) + (P_D_given_not_H * P_not_H)

# Posterior probability of having disease given a positive test
P_H_given_D = bayes_theorem(P_H, P_D_given_H, P_D)

print(f"Probability of having disease after positive test: {P_H_given_D:.4f}")

```

```python
#12.Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import scipy.stats as stats

data = [[30, 10],
        [20, 20]]

# Perform chi-square test
chi2, p, dof, expected = stats.chi2_contingency(data)

# Results
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(np.round(expected, 2))

# Interpretation
alpha = 0.05
if p < alpha:
    print("Significant association (reject H0)")
else:
    print("No significant association (fail to reject H0)")
```

```python
#14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

observed = [25, 30, 20, 25]

expected = [25, 25, 25, 25]

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Significant difference (reject H0)")
else:
    print("No significant difference (fail to reject H0)")
```
