  - 1. Train the model:- The model learn from the data.
    2. Get the training loss:→ This shows how well the model fits the  
       training data.
    3. Get the validation loss:→ This shows how well the model works on new,
       unseen data.
    4. Compare the two losses:

      - Both losses are low and close:-Good model
      
    - Training loss is low, but validation loss is high:

      Model is overfitting (memorizing instead of learning).

    - Both losses are high:

      Model is underfitting (not learning enough).

#6.How do we handle categorical variables in Machine Learning? What are the common techniques?
   - In Machine Learning, handling categorical variables is essential because
     most models require numerical input. Categorical variables contain labels or names instead of numbers (e.g., "Male", "Female", "Red", "Green").

     Common Techniques:
     1. One-Hot Encoding-> Create a new binary feature for each category.
     2. Lable Encoding-> Assign a unique integer value to each category.
     3. Ordinal Encoding-> Assign a numerical value to each category based on
        its order or ranking.
     4. Binary Encoding-> Converts categorical variables into binary codes.
     5. Hashing-> Uses a hash function to map categorical values to numerrial
        values.

#15.What is causation? Explain difference between correlation and causation with an example.
   - Causation refers to a relationship where one variable directly affects   or causes a change in another.In simple terms, if A causes B, then a change in A will produce a change in B.
  - Correlation means two things happen together, but one doesn’t necessarily
   cause the other.
   Example:Ice cream sales and drowning cases both go up in the summer.
   They're correlated because they increase at the same time — but eating ice cream doesn't cause drowning.
  - Causation means one thing directly causes the other.
   Example:Smoking causes lung disease.
   This is causation — there's scientific proof that smoking directly leads to health problems.
   
#16.What is an Optimizer? What are different types of optimizers? Explain each with an example.
   - An optimizer is an algorithm used in machine learning and deep learning
     to adjust model parameters (like weights) in order to minimize error (loss) and improve performance.

     There are several types of optimzer:

     1.Gradient Descent (GD):Use the entire dataset to compute the gradient descent and updates parameter in the direction that reduce the loss.
     e.g:Minimize the cost of predicting housing prices.

     2.Stochastic Gradient Descent (SGD):Updates the model using one sample at a time.e.g->Online learning or streaming data.

     3.Mini-Batch Gradient Descent:A middle ground between GD and SGDand
       Updates using a small batch of data.e.g->Standard in deep learning (fast and stable).etc.

#21.What is feature scaling? How does it help in Machine Learning?
   - Feature scaling is the process of adjusting the range of input features
     (columns in your data) so they are on the same scale — usually between 0 and 1 or with mean = 0 and standard deviation = 1.

     Feature scaling help in Machine learning:
     
     - Speeds up training (especially for gradient-based models).
     - improves model accuracy and stability.
     - Ensures equal weight for all features.
     - Essential for distance-based models like KNN and K-Means.
