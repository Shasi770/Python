```python
#2.Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
data = {'Name': ['Alice', 'Charlie', 'David'],
                   'Age': [25, 30, 35],
                   'Salary': [50000, 60000, 70000]}
df = pd.DataFrame(data)
mean_salary = df['Salary'].mean()
print(f"mean of salary column: {mean_salary}")
```

```python
#3.Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.scatter(x, y, color = 'red', marker = 'o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {'math': [80, 75, 85, 70, 90],
        'physics': [90, 85, 78, 92, 88],
        'chemistry': [70, 80, 90, 82, 75]}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',  fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#5.Generate a bar plot using Plotly.
import plotly.express as px
data = {'Name': ['Alice', 'Charlie', 'David'],
        'Salary': [500, 600, 700]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Name', y='Salary', title='Bar Plot')
fig.show()

```

```python
#7.Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np
array1 = np.array([1, 2, 3],)

array2 = np.array([4, 5, 6])
result = np.multiply(array1, array2)

print(result)
```

```python
#8.Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [10, 12, 14, 16, 18]
y2 = [8, 10, 12, 14, 16]

plt.plot(x, y1, label='Line 1 - Growth', color='blue', marker='o')
plt.plot(x, y2, label='Line 2 - Target', color='green', linestyle='--', marker='x')

plt.title('Line Plot with Multiple Lines')
plt.xlabel('Time')
plt.ylabel('Value')

plt.legend()

plt.grid(True)
plt.show()
```

```python
#9.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd
data = {'Name': ['Alice', 'Charlie', 'David'],
        'Age': [25, 30, 35],
        'Salary': [50000, 60000, 70000]}

df = pd.DataFrame(data)
threshold = 30
filtered_df = df[df['Age'] > threshold]
print(filtered_df)

```

```python
#10.Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt
data = [8, 10, 12, 12, 16, 18, 18, 20, 20, 20, 23, 25, 25]
sns.histplot(data, bins=7, kde=True, color = 'blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of data')
plt.show()
```
