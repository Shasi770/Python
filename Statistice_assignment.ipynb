#10.Explain nominal, ordinal, interval, and ratio levels of measurement.

     any inherent or natural order that is called as nominal data (no order in categorical data).

#14.What is variance, how is it calculated.
   - The average of squared differences from the mean valve.

     Steps to Calculate Variance:-

     1.Find the mean of the dataset.

     2.Subtract the mean from each data point (find the deviation).

     3.Square each deviation.

     4.Average the squared deviations (divide by N for population, n−1 for
      sample).

#15.What is standard deviation, and why is it important?
   - Standard deviation is a statistical measure that quantifies the amount of
     variation or dispersion in a set of data values. It is the square root of the variance, providing a measure of how spread out the values in a dataset are in the same units as the data itself.

   - Standard deviation is important because it tells us how spread out or
     consistent the data is around the mean (average). It's a key measure in statistics for understanding variability, risk, and reliability of data.

#16.Define and explain the term range in statistics.
   - In statistics, the range is a measure of dispersion that indicates the
     spread of values in a dataset.

   - The range is calculated as the difference between the maximum and minimum
     values in the dataset.

     Range = Maximum Value - Minimum Value
     Example:
     For the dataset: [3, 7, 10, 15, 18]

     Maximum = 18

     Minimum = 3

     Range = 18 - 3 = 15

#17.What is the difference between variance and standard deviation?
   - Standard deviation is the spread of a group of numbers from the mean. The
     variance measures the average degree to which each point differs from the mean.

#18.What is skewness in a dataset?

#22.What does correlation measure in statistics?

```python
#4.Implement sampling techniques like random sampling and stratified sampling.
import pandas as pd
from sklearn.model_selection import train_test_split

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen'],
    'Score': [85, 78, 90, 65, 95, 70, 88, 76],
    'Group': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B']
}

df = pd.DataFrame(data)
print(df)

random_sample =  df.sample(frac=0.5, random_state=42)
print("\nRandom Sample:")
print(random_sample)

stratified_sample, _ = train_test_split(df, test_size=0.5, stratify=df['Group'], random_state=42)
print("\nStratified Sample (based on Group):")
print(stratified_sample)
```

```python
#8.Generate a dataset and demonstrate positive and negative skewness.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
positive_skew = np.random.exponential(scale=2.0, size=1000)
negative_skew = -np.random.exponential(scale=2.0, size=1000)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(positive_skew, bins=20, color='skyblue', edgecolor='black')
plt.title(f'Positive Skew\nSkewness: {skew(positive_skew):.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(negative_skew, bins=30, color='salmon', edgecolor='black')
plt.title(f'Negative Skew\nSkewness: {skew(negative_skew):.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#9.Write a Python script to calculate covariance between two datasets.
import numpy as np
from scipy.stats import norm

dataset1 = np.array([1, 2, 3, 4, 5])
dataset2 = np.array([5, 4, 3, 2, 1])

mean_x = np.mean(dataset1)
mean_y = np.mean(dataset2)

covariance = np.sum((dataset1 - mean_x) * (dataset2 - mean_y)) / (len(dataset1) - 1)

print("Covariance between two datasets:", covariance)

```

```python
#12.Implement and compare simple random sampling and systematic sampling.
import numpy as np
import random
import matplotlib.pyplot as plt

population = np.arange(1, 1001)

simple_random_sample = random.sample(list(population), 100)

k = 10
start = random.randint(0, k - 1)
systematic_sample = population[start::k][:100]

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(simple_random_sample, bins=10, color='skyblue', edgecolor='black')
plt.title("Simple Random Sampling")

plt.subplot(1, 2, 2)
plt.hist(systematic_sample, bins=10, color='salmon', edgecolor='black')
plt.title("Systematic Sampling")

plt.tight_layout()
plt.show()
```

```python
#18.Implement Z-score normalization and explain its significanc.
'''Z-score normalization, also known as standardization, is significant because
 it transforms data to have a mean of 0 and a standard deviation of 1, making
  it easier to compare and analyze data with different scales'''
import numpy as np
data = np.array([10, 20, 30, 40, 50])

mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

print("Original Data:", data)
print("Z-score Normalized Data:", z_scores)
```

```python
#19.Compare two datasets using their standard deviations.
import numpy as np

dataset1 = np.array([10, 12, 11, 13, 12, 11, 10])
dataset2 = np.array([5, 20, 15, 30, 25, 10, 35])

std_dev1 = np.std(dataset1)
std_dev2 = np.std(dataset2)

if std_dev1 > std_dev2:
    print("Dataset 1 has a higher standard deviation.")
elif std_dev1 < std_dev2:
    print("Dataset 2 has a higher standard deviation.")

else:
    print("Both datasets have the same standard deviation.")
```

```python
#20.Write a Python program to visualize covariance using a heatmap.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {'A': [5, 8, 11, 15, 18],
        'B': [7, 9, 12, 17, 21],
        'C': [9, 11, 15, 20, 25]}
df = pd.DataFrame(data)
cov_matrix = df.cov()
plt.figure(figsize=(6, 4))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Covariance Heatmap")
plt.show()
```

```python
#24.Implement the Pearson and Spearman correlation coefficients for a dataset.
import pandas as pd
from scipy.stats import pearsonr, spearmanr
data = {
    'X': [10, 20, 30, 40, 50],
    'Y': [12, 24, 33, 45, 55]
}
df = pd.DataFrame(data)
pearson_corr, _ = pearsonr(df['X'], df['Y'])
spearman_corr, _ = spearmanr(df['X'], df['Y'])

print(f"Pearson Correlation: {pearson_corr:.3f}")
print(f"Spearman Correlation: {spearman_corr:.3f}")
```
